```python
import matplotlib.pyplot as plot
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

figure = plot.figure()

axis = figure.add_subplot(111, projection='3d')

p1 = np.array([4, 2, 1])
p2 = np.array([3, 3, 1])

axis.quiver(0, 0, 0, p1[0], p1[1], p1[2], arrow_length_ratio=0.1, color='black')
axis.quiver(0, 0, 0, p2[0], p2[1], p2[2], arrow_length_ratio=0.1, color='black')

# Subtracting two position vectors to the points p1 and p2 gives us the red vector between the position vectors terminal points
axis.quiver(p1[0], p1[1], p1[2], (p2[0] - p1[0]), (p2[1] - p1[1]), (p2[2] - p1[2]), arrow_length_ratio=0.5, color='red')


axis.set_box_aspect([1, 1, 1])

axis.set_xlim([0, 5])
axis.set_ylim([0, 5])
axis.set_zlim([0, 5])

plot.show()

```
