```python
import matplotlib.pyplot as plot
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

figure = plot.figure()
axis = figure.add_subplot(111, projection='3d')

vector = np.array([4, 2, 3])

axis.quiver(0, 0, 0, vector[0], vector[1], vector[2], arrow_length_ratio=0.1, color='blue')

axis.set_box_aspect([1, 1, 1])

axis.set_xlim([0, 5])
axis.set_ylim([0, 5])
axis.set_zlim([0, 5])

axis.text(0.5, 0, 0, "x-axis", color='red')
axis.text(0, 0.5, 0, "y-axis", color='green')
axis.text(0, 0, 0.5, "z-axis", color='blue')

axis.set_box_aspect([1, 1, 1])

# axis.view_init(elev=20, azim=120)

plot.show()
```
